```python
import tkinter as tk
from tkinter import messagebox

def subtri():
    def calcular():
        try:
            vlprod = float(entrada_valor_produto.get())
            if var_outr_vl.get() == "S":
                vl_desp_int = float(entrada_despesas.get())
                vl_frete_int = float(entrada_frete.get())
                vl_ipi_int = float(entrada_ipi.get())
                vl_seguro_int = float(entrada_seguro.get())
                vl_desconto_int = float(entrada_desconto.get())
                base_produtos = vlprod + vl_desp_int + vl_ipi_int + vl_seguro_int - vl_desconto_int
            else:
                base_produtos = vlprod

            operacao = var_operacao.get()
            if operacao == "I":
                mva_original_int = float(entrada_mva_original.get())
                aliquota_int = float(entrada_aliquota.get())
                vl_icms_int = float(entrada_icms.get())
                base_st = base_produtos * (1 + (mva_original_int / 100))
                icms_st = (base_st * (aliquota_int / 100)) - vl_icms_int
            elif operacao == "E":
                mva_ajustada_int = float(entrada_mva_ajustada.get())
                aliquota_int = float(entrada_aliquota.get())
                vl_icms_int = float(entrada_icms.get())
                base_st = base_produtos * (1 + (mva_ajustada_int / 100))
                icms_st = (base_st * (aliquota_int / 100)) - vl_icms_int
            else:
                raise ValueError("Operação inválida")

            resultado_texto.set(f"NCM: {entrada_ncm.get()}\nO valor da Base de Cálculo de ST é: {base_st}\nO valor de ST a recolher é: {icms_st}")
        except ValueError as e:
            messagebox.showerror("Erro", str(e))

    janela = tk.Tk()
    janela.title("Cálculo de Substituição Tributária")

    tk.Label(janela, text="NCM dos produtos:").grid(row=0, column=0)
    entrada_ncm = tk.Entry(janela)
    entrada_ncm.grid(row=0, column=1)

    tk.Label(janela, text="Valor dos produtos (R$):").grid(row=1, column=0)
    entrada_valor_produto = tk.Entry(janela)
    entrada_valor_produto.grid(row=1, column=1)

    var_outr_vl = tk.StringVar(value="N")
    tk.Label(janela, text="Existem outros valores (S/N)?").grid(row=2, column=0)
    tk.Radiobutton(janela, text="Sim", variable=var_outr_vl, value="S").grid(row=2, column=1)
    tk.Radiobutton(janela, text="Não", variable=var_outr_vl, value="N").grid(row=2, column=2)

    tk.Label(janela, text="Despesas Acessórias:").grid(row=3, column=0)
    entrada_despesas = tk.Entry(janela)
    entrada_despesas.grid(row=3, column=1)

    tk.Label(janela, text="Frete:").grid(row=4, column=0)
    entrada_frete = tk.Entry(janela)
    entrada_frete.grid(row=4, column=1)

    tk.Label(janela, text="Valor do IPI:").grid(row=5, column=0)
    entrada_ipi = tk.Entry(janela)
    entrada_ipi.grid(row=5, column=1)

    tk.Label(janela, text="Valor do Seguro:").grid(row=6, column=0)
    entrada_seguro = tk.Entry(janela)
    entrada_seguro.grid(row=6, column=1)

    tk.Label(janela, text="Valor do Desconto:").grid(row=7, column=0)
    entrada_desconto = tk.Entry(janela)
    entrada_desconto.grid(row=7, column=1)

    var_operacao = tk.StringVar(value="I")
    tk.Label(janela, text="Tipo de Operação (I/E):").grid(row=8, column=0)
    tk.Radiobutton(janela, text="Interna", variable=var_operacao, value="I").grid(row=8, column=1)
    tk.Radiobutton(janela, text="Interestadual", variable=var_operacao, value="E").grid(row=8, column=2)

    tk.Label(janela, text="MVA Original:").grid(row=9, column=0)
    entrada_mva_original = tk.Entry(janela)
    entrada_mva_original.grid(row=9, column=1)

    tk.Label(janela, text="MVA Ajustada:").grid(row=10, column=0)
    entrada_mva_ajustada = tk.Entry(janela)
    entrada_mva_ajustada.grid(row=10, column=1)

    tk.Label(janela, text="Alíquota interna de ICMS:").grid(row=11, column=0)
    entrada_aliquota = tk.Entry(janela)
    entrada_aliquota.grid(row=11, column=1)

    tk.Label(janela, text="Valor do ICMS dos produtos:").grid(row=12, column=0)
    entrada_icms = tk.Entry(janela)
    entrada_icms.grid(row=12, column=1)

    tk.Button(janela, text="Calcular", command=calcular).grid(row=13, column=0, columnspan=2)

    resultado_texto = tk.StringVar()
    tk.Label(janela, textvariable=resultado_texto).grid(row=14, column=0, columnspan=2)

    janela.mainloop()

subtri()
```
